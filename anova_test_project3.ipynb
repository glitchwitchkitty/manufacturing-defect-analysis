```python
import pandas as pd
from scipy.stats import f_oneway

df = pd.read_csv("anova_defects_per_week.csv")

df.columns
df.head()
```

```python
df = df.rename(columns={"Count of Defect Type": "Defect_Count"})
```

```python
g1 = df[df["Production Model"].str.contains("595130-195")]["Defect_Count"]
g2 = df[df["Production Model"].str.contains("595214-125")]["Defect_Count"]
g3 = df[df["Production Model"].str.contains("595242-854")]["Defect_Count"]

len(g1), len(g2), len(g3)
```

```python
f_stat, p_value = f_oneway(g1, g2, g3)
f_stat, p_value
```

```python
alpha = 0.05
if p_value <= alpha:
    decision = "Reject H0"
else:
    decision = "Fail to reject H0"
decision
```
